*1.2 Differences between how a Classification Decision Tree makes  predictions vs. how Multiple Linear Regression makes predictions*

Difference btween classificaiton and regerssion: 
- Classification: predicts categorial outcomes such as "yes/no", "true or false", or even "class a/class b/ class c".
- Regression: predicts continuos numerical outcomes. 


How Classification Decision Tree makes  predictions
1. Starting at a root node: the  root represents the entire data set and a decision is made depeding on the feature that will best split the data into different groups. 
2. Splitting data: from the root, several decisions (nodes) are obtained. The tree evaluates one feature and its threshold to split the dataset into subsets that are more homogenous in terms of class labels. 
3. Following branches: the process of diviing data into subsets continues adown the tree, each branch leads to tighter and more refined subset of data, until a leaf node is reached. 
4. Assigning a class: the leaf node represents the final prediciton, which is often determined by the most common class in that node. 

The process that a CDT follows is sequential and hierarchical, so the decisions at the top are the ones that include more data, while the decisions at the bottom are more specific and smaller. Decisions are made one step at a time, which narrows the possibilites of classes. The tree's construction is often measured using Gini Impurtiy or Entropy classificaiton. 

High level Nture of a deciion tree:
- Each node is a decision point based on a feature. 
- Each branch represents the outcome of a decision. 
- The leaf nodes contain the final predictions. 


How Mutliple Linear Regression makes predictions
- Multiple linear regression makes predicitions using linear combinations. 
1. Each feature is weighted by a coefficient whihc indicates its contribution to the prediciton. 
2. The prediction is a continuos value calculated with the formula  Prediction = $β_0 + β_1·X + β_2·X + ... + β_n ·X_n$
3. This model minimizes the prediciton error (represented by a $R^2$) to determined the best coefficients


The major difference between these two modesl is that regression uses mathematical formula to make its prediction, while the other uses a rule based sequential splitting.

*4.2 Code to "train" a classification tree clf using only the List Price variable*
- The following code demonstrates how to train the Decision Tree Classifier using only the List Price variable to predict whether a book is hardcover or paperback.
- The maximum tree depth is 2. 

```python
# Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Load the dataset (ensure ab_reduced_noNaN is already processed)
# Perform an 80/20 split
random_seed = 42
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(
    ab_reduced_noNaN, test_size=0.2, random_state=random_seed
)

# Prepare data for training
y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])["H"]
X = ab_reduced_noNaN_train[["List Price"]]

# Initialize and train the DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=2, random_state=random_seed)
clf.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(10, 6))
tree.plot_tree(clf, feature_names=["List Price"], class_names=["Paperback", "Hardcover"], filled=True)
plt.title("Decision Tree: Predicting Hardcover or Paperback")
plt.show()

# Display number of observations in training and test datasets
print(f"Training set observations: {len(ab_reduced_noNaN_train)}")
print(f"Test set observations: {len(ab_reduced_noNaN_test)}")
```

4. At the leaf node, the class with the highest probability is the final prediction.

- Intermediate Nodes: these nodes refine the classification for the subsets of data. Further split the data using additional conditions based on the feature. 
- Leaf nodes: the terminal nodes. Each node has a prediction class, the proportion of samples belonging to each class in that node, and the total number of samples in that node.

```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Step 2: Load and Preprocess the Dataset
# Replace 'file_path' with the correct dataset file path if loaded locally
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Preprocessing the dataset
ab_reduced = ab.drop(columns=["Weight_oz", "Width", "Height"], errors='ignore')  # Remove specified columns
ab_reduced_noNaN = ab_reduced.dropna()  # Drop rows with NaN values
ab_reduced_noNaN.loc[:, "Pub year"] = ab_reduced_noNaN["Pub year"].astype(int)
ab_reduced_noNaN.loc[:, "NumPages"] = ab_reduced_noNaN["NumPages"].astype(int)
ab_reduced_noNaN.loc[:, "Hard_or_Paper"] = ab_reduced_noNaN["Hard_or_Paper"].astype("category")

# Step 3: Split Dataset into Train/Test Sets
random_seed = 42
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(
    ab_reduced_noNaN, test_size=0.2, random_state=random_seed
)

# Step 4: Train the Decision Tree Classifier
X2 = ab_reduced_noNaN_train[["NumPages", "Thick", "List Price"]]  # Predictors
y2 = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])["H"]  # Target

clf2 = DecisionTreeClassifier(max_depth=4, random_state=random_seed)
clf2.fit(X2, y2)

# Step 5: Visualize the Decision Tree
plt.figure(figsize=(20, 10))
plot_tree(
    clf2, 
    feature_names=["NumPages", "Thick", "List Price"], 
    class_names=["Paperback", "Hardcover"], 
    filled=True, 
    rounded=True
)
plt.title("Decision Tree for clf2: Predicting Hardcover vs Paperback")
plt.show()
```

```python
#6.1 Confusion Matrix for clf
# Import necessary libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score
import matplotlib.pyplot as plt

# Prepare training and test data for clf
y_train_clf = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])["H"]
X_train_clf = ab_reduced_noNaN_train[["List Price"]]
y_test_clf = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])["H"]
X_test_clf = ab_reduced_noNaN_test[["List Price"]]

# Train clf
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train_clf, y_train_clf)

# Predict using clf
y_pred_clf = clf.predict(X_test_clf)

# Confusion Matrix for clf
cm_clf = confusion_matrix(y_test_clf, y_pred_clf)

# Visualize confusion matrix for clf
ConfusionMatrixDisplay(cm_clf, display_labels=["Paperback", "Hardcover"]).plot(cmap="Blues")
plt.title("Confusion Matrix for clf")
plt.show()

# Metrics for clf
metrics_clf = {
    "Accuracy": accuracy_score(y_test_clf, y_pred_clf),
    "Sensitivity": cm_clf[1, 1] / (cm_clf[1, 1] + cm_clf[1, 0]),  # TP / (TP + FN)
    "Specificity": cm_clf[0, 0] / (cm_clf[0, 0] + cm_clf[0, 1])   # TN / (TN + FP)
}

print("Metrics for clf:")
print({k: round(v, 3) for k, v in metrics_clf.items()})
```

4. Right bottom corner: True Positive = Predicted hardcover is actually a hardcover.

In linear regression, coefficients directly quantify the change in the response variable for a one-unit change in the predictor variable, assuming all other variables are held constant, making interpretation straightforward and grounded in linear relationships. In contrast, feature importances in decision trees are derived from the overall reduction in impurity (e.g., Gini index or entropy) across all splits involving a feature, which provides a global measure of how influential a variable is for predictions but does not specify the direction or magnitude of its effect on the response. Thus, while regression coefficients offer interpretable numerical effects, feature importances reflect a broader, less direct influence on model performance.
