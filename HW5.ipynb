**Question 5**

Understanding the info:
- Data about a group of couples and their likelihood to tilt their heads while kissing
    - 80 out of 124 couples tilt their head 
        - That is 64.5%
- We will using a 50/50 coin flipping model to calculate the p-value         

Original Hypothesis:
- Null (H0): couples don't have left or right head tilt tendecies when kissing 
- Alternative (Ha): couples do have head tilt tendencies when kissing


Relevant info:
- n=124 couples
- Observed results for head tilt while kissing: 64.5% 

Hypothesis using coin flipping model:
- Null: All couples have a 50% chance of tilting the head to the right and 50% of tilting it to the left

```python
#Simulation for p-value

import numpy as np

n = 124  #
observed_proportion = 80 / 124  
p_null = 0.5  


n_simulations = 10000  
np.random.seed(42)  

simulated_counts = np.random.binomial(n, p_null, n_simulations)
simulated_proportions = simulated_counts / n

p_value = np.mean(simulated_proportions >= observed_proportion)
p_value
```

The p-value for our null hypothesis is 0.001, which is an exteremely low value. This value represents that there is 0.001 probability of our null hypothesis to be true under the data we are using. Using this evidence we conclude that couples DO have a preference of tilting their head when kissing. Therefore we say there is enough evidene to reject the null hypothesis. Our arguments being:
1. 0.001>=p is very strong evidnece against the null hypothesis
2. 64.5% of our whole population (couples) tilting their head either to the right or left when kissing, which is more than half of the entire data.

```python
#Code computing a p-value for a one sided hypothesis test for the Vaccine Data Analysis:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the random seed for reproducibility
np.random.seed(42)

# Creating the DataFrame
data = {
    "PatientID": range(1, 11),
    "Age": [45, 34, 29, 52, 37, 41, 33, 48, 26, 39],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F"],
    "InitialHealthScore": [84, 78, 83, 81, 81, 80, 79, 85, 76, 83],
    "FinalHealthScore": [86, 86, 80, 86, 84, 86, 86, 82, 83, 84],
}

df = pd.DataFrame(data)

# Plotting the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['InitialHealthScore', 'FinalHealthScore']])
plt.title('Comparison of Initial and Final Health Scores')
plt.ylabel('Health Score')
plt.xticks([0, 1], ['Initial Health Score', 'Final Health Score'])
plt.grid(axis='y')
plt.show()

# Calculating the score difference
df['ScoreDifference'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Bootstrapping
n_iterations = 1000
boot_means = []
for _ in range(n_iterations):
    boot_sample = np.random.choice(df['ScoreDifference'], size=len(df), replace=True)
    boot_means.append(np.mean(boot_sample))

# Observed mean difference
observed_mean_diff = np.mean(df['ScoreDifference'])

# Confidence interval
conf_interval = np.percentile(boot_means, [2.5, 97.5])

# One-sided p-value calculation
# Testing if the final score is significantly greater than the initial score
p_value_one_sided = np.sum(np.array(boot_means) >= observed_mean_diff) / n_iterations

# Output results
print(f'Observed Mean Difference: {observed_mean_diff}')
print(f'95% Confidence Interval (two-sided): {conf_interval}')
print(f'One-sided p-value: {p_value_one_sided}')

    
```

*Major changes in this new code:*
- Addition of a one-sided p-value calculation: p_value_one_sided = np.sum(np.array(boot_means) >= observed_mean_diff) / n_iterations
    - This is calculated by first establishing the proportion of the bootstrapped means which are more or equal to the observed mean difference. 
    - By finding this, the code assumes that the final health score is greater than the initial one, and gives as a result a one sided p-value    
    
*Change in interpretation of hypothesis test:*

While the two sided coded perfomed bootstrapping and provided a 95% confidence interval, wihtout really defining a hypothesis. The new code has set the focus to be to prove if the Final Health Score is greater than the Initial Health Score.

**Question 8:**

*Problem Introduction:*
    
On this question, the main purpose is to provide a statistical analysis of the experiment performed in STA130, where 80 student were given the task to taste one cup of tea and tell whether they think the milk or the tea was poured first. The results gave that 49 students are able to correctly state which was poured first, and we will be analyzing those results, and determine whether the recognition was due to their ability to tell them apart, or by pure guess. 
This experiment is similar to the original experiment done by Fisher and Bristol. It is really the same experiment, with the same purpose, yet with a higher population sample. In the original experiment, Bristol cliamed she could taste the difference of when the tea or the milk was poured first. So, Fisher tested Bristol by giving her 8 cups. The result was that Bristol identified all of the 8 cups correctly. 
- Similarities: the task being perfomed is the same, the purpose and focus of hypotheis is also the same. 
- Differences: Fisher only tested Bristol's ability, while STA130 is testing 80 students. Additionally, Fisher was based Bristol's taste, while STA130 is focusing on a general population. 

*Null hypothesis Statement:* 
- Formal: 
H0: p=0.5, where p represents the proportion of students who are able to correctly identify the elemtn poured in first.
- Informal:
the students in the STA130 experiment were only guessing, so half are expected to get the answer right 

*Alternative Hypothesis Statement:*
- Formal: 
Ha: p>0.5, where the p value implies that the studetns are more likely to identify the order than what is expected only by guessing
- Informal: 
more than half of the studnets will identify the element poured in first. 

*Quantitative Analysis:*

Using a sample proportion, and knowing: 
- sample size (n) = 80 
- number of recognition of element in tea (x)=49
Then the sample proportion is p=49/80, which results in roughts 0.6125. 

With this result, we can now perfrom a hypothesis test. Since we have the observed sample mean: 0.6125, the mean under null hypothesis: 0.5, and the n value: 80. 

```python
#Methodology Code and Explanations: 
import numpy as np
import scipy.stats as stats

# Set seed for reproducibility
import numpy as np
import scipy.stats as stats

# Set seed for reproducibility
np.random.seed(42)

# Given values
n = 80
x = 49
p_hat = x / n
p_0 = 0.5

# Calculate the standard error
std_error = np.sqrt((p_0 * (1 - p_0)) / n)

# Calculate the z-score
z_score = (p_hat - p_0) / std_error

# Calculate the p-value (one-tailed test)
p_value = 1 - stats.norm.cdf(z_score)

z_score, p_value
```

In here we can recognize that the p-value is: 0.0220...

```python
#Visual presentation
# Sampling distribution plot
x_vals = np.linspace(0, 1, 1000)
sampling_dist = stats.norm.pdf(x_vals, loc=p_0, scale=std_error)

plt.figure(figsize=(10, 5))

# Plot the sampling distribution
plt.plot(x_vals, sampling_dist, label='Null Hypothesis Sampling Distribution', color='blue')
plt.axvline(p_hat, color='red', linestyle='--', label=f'Observed Proportion: {p_hat:.3f}')
plt.fill_between(x_vals, sampling_dist, where=(x_vals >= p_hat), color='red', alpha=0.3, label=f'p-value Region (p = {p_value:.3f})')

plt.title('Sampling Distribution of Proportions (Null Hypothesis)')
plt.xlabel('Sample Proportion')
plt.ylabel('Probability Density')
plt.legend()

plt.show()

# Z-distribution plot
z_vals = np.linspace(-3, 3, 1000)
z_dist = stats.norm.pdf(z_vals)

plt.figure(figsize=(10, 5))

# Plot the standard normal distribution
plt.plot(z_vals, z_dist, label='Z-Distribution', color='blue')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score: {z_score:.2f}')
plt.fill_between(z_vals, z_dist, where=(z_vals >= z_score), color='red', alpha=0.3, label=f'p-value Region (p = {p_value:.3f})')

plt.title('Z-Distribution for Hypothesis Test')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend()

plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given values
n = 80  # sample size
x = 49  # number of correct identifications
p_hat = x / n  # sample proportion
p_0 = 0.5  # hypothesized proportion under the null

# Confidence level
confidence_level = 0.95
z_alpha_over_2 = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # z-score for 95% confidence level

# Calculate the margin of error
std_error = np.sqrt((p_hat * (1 - p_hat)) / n)
margin_of_error = z_alpha_over_2 * std_error

# Calculate the z-score and p-value for the test
z_score = (p_hat - p_0) / np.sqrt((p_0 * (1 - p_0)) / n)
p_value = 1 - stats.norm.cdf(z_score)

# Set up the plot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot confidence interval for the sample proportion
ax.errorbar(x=0, y=p_hat, yerr=margin_of_error, fmt='o', color='blue', label='Sample Proportion  95% CI', capsize=5)

# Add the null hypothesis proportion as a red line
ax.axhline(y=0.5, color='red', linestyle='--', label='Null Hypothesis Proportion (p = 0.5)')

# Shade the p-value region
z_vals = np.linspace(-3, 3, 1000)
z_dist = stats.norm.pdf(z_vals)
ax2 = ax.twinx()  # Create a secondary y-axis for the p-value distribution
ax2.plot(z_vals, z_dist, color='gray', label='Z-Distribution', alpha=0.5)
ax2.fill_between(z_vals, 0, z_dist, where=(z_vals >= z_score), color='red', alpha=0.3, label=f'p-value Region (p = {p_value:.3f})')

# Add labels and title
ax.set_ylim(0.4, 0.7)
ax.set_xticks([])
ax.set_ylabel('Proportion')
ax2.set_ylabel('Z-Distribution Density')
ax.set_title('95% Confidence Interval and P-Value for Sample Proportion')

# Add legends
ax.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```
