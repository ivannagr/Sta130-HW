*1.1 Difference between Simple and Multiple Linear Regression*
- Simple models relationship between an outcome variable Y and a single predictor Y: $Y$ = $β_0$ + $β_1X$ + $ϵ$
    - $β_0$ is the intercept and $β_1$ is the slope of $X$ 
    
- Multiple extends the simple model to multiple predictors by incorporating more variables:  $Y$ = $β_0$ + $β_1X_1$ + $β_2X_2$ ... $β_nX_n$ + $ϵ$
    - This accounts for combined effect of several predictors, which increases the accuracy of the predictions and it also provides a better understanding of the relationship, is mostly useful for complex data 
    
*1.2 Continus vs Indicator Variable in Simple Linear Regression*
- Continuos variable predictor captures a linear trend over a range of values. 
    - For example: $Y$ = $β_0$ + $β_1 *$ Age + $ϵ$
- Indicator variable is a binary variable, meaning its values are either 0 or 1, and it represents categories. Shifting the intercept by $β_1$ when the indicator is 1, wihtout a slope effect in order to compare groups 
    - For example: $Y$ = $β_0$ + $β_1 *$ 1(Female) + $ϵ$
    
*1.3 Introducing Indicator Variable with Continuos Variable in Multiple Linear Regression*

Adding an indicator varialbe with a continuos variable adjusts the outcome for group differences. In order to allow different intercepts for each group while keeping the slope ($β_1$) constant, and capturing different baseline levels for $Y$ by group 
- For example: $Y$ = $β_0$ + $β_1X$ + + $β_2 *$ 1(Group) + $ϵ$
    - $X$ is continuos 
    - 1(Group) is an indicator 
    - This models $Y$ with an adjusted intercept for each group:
        - 1(Group) = 0, $Y$= $β_0$ + $β_1X$
        - 1(Group) = 1, $Y$= ($β_0$ +$β_2$)+ $β_1X$

*1.4 Adding an Interaction between Continuos and Indicator Variable*

Adding a continuos variable and an indicator variable allows differnt slopes for different groups 
- For example: $Y$ = $β_0$ + $β_1X$ + + $β_2 *$ 1(Group) + $β_3 *$ ($X *$ 1(Group)) + $ϵ$
    - 1(Group) = 0, $Y$= $β_0$ + $β_1X$
    - 1(Group) = 1, $Y$= ($β_0$ +$β_2$)+($β_1$ +$β_3$)$X$
        - $β_3$ allows the slope of $X$ to vary by group, which gives flexibility to model scenarios in which the relationship between $Y$ and $X$ varies

*1.5 Multiple Linear Regression with Only Indicator Variables from Non-binary Categorical Varialbe*

For a non-binary categorical varialbe with $k$ categories, this model uses $k-1$ indicator variables, leaving 1 as the baseline. 
- For example, a categorical variable Region with values {Norht, South, East, West}, North can be used as the baseline, while {South, East, West} as the indicators: 
    - $Y$ = $β_0$ + $β_1 *$ 1(South) + + $β_2 *$ 1(East) + $β_3 *$ 1(West) + $ϵ$

    - Each coefficient represents the shift in $Y$ from the baseline, and using $k-1$ ensures the impact of each category is different, which avoids the redundancy. So to model the impact of region we use the following forms: 
        - Basline (North): $Y$ = $β_0$ 
        - South:$Y$ = $β_0$ + $β_1$
        - East:$Y$ = $β_0$ + $β_2$
        - West:$Y$ = $β_0$ + $β_3$

*3.1 Additive Specification*
The model assumes each predictor variable contributes to the outcome without any interaction effect. 
- Continuos variable: age
- Binary variable: Gender, female = 0 and male = 1
- Categorical variable: social activity level, either low, medium or high 
- Model: $Y$ = $β_0$ + $β_1$(Age) + $β_2$(Gender) + $β_3$(Social Activity Level) 
    - Age ($β_1$) quantifies the effect of age on the outcome, postive $β_1$ indicates as age increases, the level of outcome also increases
    - Gender ($β_2$) indicates the difference in outcome between genders. If $β_2$ is positive, males have a higher value of the outcome than females 
    - soical Activity Level ($β_3$) represents the effect of differing levels of social activity, and each level has an effect size compared to the baseline 


```python
#3.1 Simulated data: 
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Simulating data for CSCS
np.random.seed(42)
n = 200

# Continuous variable: Age (20-80 years)
age = np.random.randint(20, 80, n)

# Binary variable: Gender (0 = Female, 1 = Male)
gender = np.random.choice([0, 1], size=n)

# Categorical variable: Social activity level (Low, Medium, High)
social_activity_level = np.random.choice(['Low', 'Medium', 'High'], size=n)

# Continuous Outcome variable (some arbitrary linear function + noise)
outcome = 50 + 0.5 * age + 3 * gender + np.random.normal(0, 10, n)

# Create a DataFrame
df = pd.DataFrame({
    'Age': age,
    'Gender': gender,
    'SocialActivityLevel': social_activity_level,
    'Outcome': outcome
})
```

*3.2 Synergistic Specification*
For the synergistic specification, we add interaction terms to model potential relationships where 2 variables combine and impact the outcome differently than they would individually. 
- Model: $Y$ = $β_0$ + $β_1$(Age) + $β_2$(Gender) + $β_3$(Social Activity Level) + $β_4$(Age · Gender)
    - $β_4$ captures interaction between age and gender, which will give the effect of age on the outcome depdening on the gender. 


```python
#3.2 Synerstifc Specification 
# Define additive model formula (no interaction terms)
additive_formula = 'Outcome ~ Age + Gender + C(SocialActivityLevel)'

# Fit the additive model
additive_model = ols(additive_formula, data=df).fit()

# Print summary for interpretation of statistical evidence
print(additive_model.summary())
```

```python
# Define interaction model formula with Age * Gender interaction term
interaction_formula = 'Outcome ~ Age * Gender + C(SocialActivityLevel)'

# Fit the interaction model
interaction_model = ols(interaction_formula, data=df).fit()

# Print summary for interpretation of statistical evidence
print(interaction_model.summary())
```

*3.3 Statistical Evidence for Predictors*
For each predictor, the model outputs coefficients, the p-vlues and the confidence interval. 
- p-values: a low p-value will indicate predictor is stastically significantly in influencing the outcome.
- CI: a narrow CI suggests precision in estimation, and if CI does not contain 0, it suggest the predictor has a meaningufl effect.
In the synergistic model, if the interaction term's p-value is low then it gives evdence that interaction between age and gender significantly influences the outcome. If the interaction is not significant, then a simpler model would be fit. 


```python
#3.3 Statistical evidence
# Generate predictions from both models
df['Additive_Pred'] = additive_model.predict(df)
df['Interaction_Pred'] = interaction_model.predict(df)
print(df[['Age', 'Gender', 'SocialActivityLevel', 'Outcome', 'Additive_Pred', 'Interaction_Pred']].head())
```

*3.4 Visualization with Pltoly*
- Additive: will show separate fit lines for each gender group without consdiering any interaction. The best fit lines for each gender are parallel, which indicate that the effect on the outcome is constant across genders. The evidence from the p-value indicate that predictors are significant individually. 
- Synergistic: display the interaction effect by shwoing different slopes for each gender if there is significant interaction between age and gender. This model includes an interaction term for Age · Gender allowing the relationship between age and outcome to differ based on gender. The best fit lines for each gender group may differ in slope.


```python
#3.3 Additive Model Visualization 
# Scatter plot for observed data points with predicted "best fit" lines (Additive Model)
fig_additive = px.scatter(df, x='Age', y='Outcome', color='Gender', 
                          labels={'Gender': 'Gender (0 = Female, 1 = Male)'})

# Add best fit lines for each gender in the additive model
for gender_val in [0, 1]:
    gender_subset = df[df['Gender'] == gender_val]
    fig_additive.add_trace(go.Scatter(x=gender_subset['Age'], 
                                      y=gender_subset['Additive_Pred'],
                                      mode='lines', 
                                      name=f'Additive Best Fit (Gender {gender_val})'))

# Update layout
fig_additive.update_layout(title="Additive Model: Outcome vs. Age with Gender as Binary Indicator")
fig_additive.show(renderer="png")
```

```python
#3.4 Interaction Model Visualization 
# Scatter plot for observed data points with predicted "best fit" lines (Interaction Model)
fig_interaction = px.scatter(df, x='Age', y='Outcome', color='Gender', 
                             labels={'Gender': 'Gender (0 = Female, 1 = Male)'})

# Add best fit lines for each gender in the interaction model
for gender_val in [0, 1]:
    gender_subset = df[df['Gender'] == gender_val]
    fig_interaction.add_trace(go.Scatter(x=gender_subset['Age'], 
                                         y=gender_subset['Interaction_Pred'],
                                         mode='lines', 
                                         name=f'Interaction Best Fit (Gender {gender_val})'))

# Update layout
fig_interaction.update_layout(title="Interaction Model: Outcome vs. Age with Gender as Binary Indicator")
fig_interaction.show(renderer="png")
```

*3.5 Interpretation of Results*
Using additive_model.summary() and interaction_model.summary(), the interpretation of statistical evidence is the following: 
- Coefficient estimates: coefficients indicate the effect of each predictor on the outcome. Age has a coefficient of 0.5, so fo each one unit increase in Age, the outcome increses by 0.5 units.
- P-values: each p-value indicates the significance of the predictor. Typically a p-value <0.05 suggests the predictor has a statiscally significant impact on the outcome. 
- Interaction Term: the (age · gender) intereaction term tests is whether the effect on the outcome depends on gender. Added to that, a low p-value for this term suggests the interaction is necessary, and this would make evident the differing slopes of the best fit lines in the interaction plot

```python
#Cell 1
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
```

*Cell 1*
Dataset pokeaman is split into two subsets, the training set and the test set: 
- Training set: 50% of the data which is used to train the models.
- Test set: the remaining 50% of the data which is used for testing model generalizability. 
The train_test_split is used to ensure that there is randomness in the split. The main goal of this setupt is to create a fair environment (50:50) to assess model generalizability, which helps to ensure there is no leakage from the test set into the training process

```python
#Cell 2 
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
```

*Cell 2: Simple Linear Model (Model 3)*
A simple linear model regression is created using smf.ols, in which the dependt varialbe is HP and the predictors are Attack and Defnse. This is a model trained on the training set pokeaman_train. 
model3_fit.summary() provides the following info:
- In sample $R^2$ which indicates how much variane of HP the model explains on thie set, coefficients, p-values, residual errors, among other statistcs.

```python
#Cell 3
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
```

*Cell 3: Generalizability of Model 3*
Predictions are generated for the test set using model3_fit.predict(). 
- In-sample R-squared: Directly from model3_fit.rsquared.
- Out-of-sample R-squared: Computed as the square of the correlation between the actual (y) and predicted (yhat_model3) values in the test set.

This measures how well the model generalizes to unseen data. If in-sample R-squared is significantly higher than out-of-sample R-squared, the model is overfitting. A small gap between the two indicates good generalizability.

The design matrix (model4_spec.exog) is a representation of the predictor variables, created by expanding the specified linear model formula into a matrix of numerical columns used to predict the outcome variable (model4_spec.endog). In the case of model4_linear_form, this expansion includes transformations such as interactions (*) and main effects (+) of the predictors.


*Creation of New Predictor Variables in the Design Matrix:*

- Each term in the formula (e.g., scale(center(Attack)) * scale(center(Defense))) is expanded into one or more columns of the design matrix.
- Interaction terms introduce additional columns representing the product of the involved variables (e.g., the interaction between scaled Attack and Defense becomes a new variable in the design matrix).
- When higher-order interactions are included (e.g., involving three or more variables), the number of columns increases exponentially.

*Multicollinearity:*
- Multicollinearity occurs when columns in the design matrix are highly correlated.
- This is exacerbated in models with interaction terms, especially when predictors are not appropriately scaled or are correlated to begin with.
- np.corrcoef(model4_spec.exog) reveals this multicollinearity by showing high correlations between certain columns of the design matrix.


*Condition Number and Numerical Stability:*
- The condition number (Cond. No.) quantifies the degree of multicollinearity and the numerical stability of the model.
- A high condition number indicates severe multicollinearity, making the model parameters sensitive to small changes in the data.
- Without centering and scaling, predictors with large magnitudes or variances contribute to an inflated condition number.


*Impact on Generalization::*
- Multicollinearity leads to unstable coefficient estimates, making predictions overly sensitive to minor variations in the data.
- This lack of robustness results in poor "out-of-sample" generalization, as the model struggles to make accurate predictions on new data.


*Effect of Centering and Scaling:* 
- Centering (subtracting the mean) and scaling (dividing by the standard deviation) mitigate multicollinearity by standardizing the predictors.
- While this reduces the condition number significantly, interaction terms can still introduce multicollinearity, especially when many terms are involved (as seen in model4_CS_spec).


Overall, the design matrix expands the linear model formula into numerical columns, including main effects and interactions. Multicollinearity arises when these columns are highly correlated, which inflates the condition number and destabilizes the model. This instability undermines "out-of-sample" prediction accuracy. *General Methodology:*
- Evidence-Based Refinements: Each model’s development leverages coefficient significance testing to iteratively refine predictors and interactions.
